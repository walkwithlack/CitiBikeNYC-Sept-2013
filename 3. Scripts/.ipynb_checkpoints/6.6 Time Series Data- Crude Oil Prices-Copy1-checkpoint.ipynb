### 3. Time-series analysis: decomposition

```python
# Decompose the time series using an additive model
from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(df_sub['value'], model='additive', period=365)
```

### 4. Testing for stationarity

```python

from statsmodels.tsa.stattools import adfuller # Import the adfuller() function

def dickey_fuller(timeseries): # Define the function
    # Perform the Dickey-Fuller test:
    print ('Dickey-Fuller Stationarity test:')
    test = adfuller(timeseries, autolag='AIC')
    result = pd.Series(test[0:4], index=['Test Statistic','p-value','Number of Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
       result['Critical Value (%s)'%key] = value
    print (result)

# Apply the test using the function on the time series
dickey_fuller(df_sub['value'])
```

Dickey-Fuller test results suggest that time series is stationary because:

The Test Statistic (-3.53) is less than the Critical Value at 5% (-2.86).

The p-value (0.0072) is less than 0.05, so reject the null hypothesis of a unit root (non-stationarity).

This means that data doesn't have a unit root and is considered stationary so it is suitable for time series modeling.

```python
# Check out a plot of autocorrelations

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf # Here, you import the autocorrelation and partial correlation plots

plot_acf(df_sub['value'])
plt.show()
```

### 5. Stationarizing the Federal Reserve Economic Data

```python
df_diff = df_sub - df_sub.shift(1) 
```

```python
# Remove the missing values that came about as a result of the differencing to perform Dickey-Fuller test.

df_diff.dropna(inplace = True) 
```

```python
df_diff.head()
```

```python
df_diff.columns
```

```python
# Check out what the differencing did to the time-series curve

plt.figure(figsize=(15,5), dpi=100)
plt.plot(df_diff)
```

```python
dickey_fuller(df_diff['value'])
```

Test Statistic = -14.2969
This is very negative, much smaller than all the critical values.

p-value  1.25e-26
This is basically zero, way less than 0.05, so super strong evidence to reject the null hypothesis.

Critical Values:
- 1% level: -3.4323
- 5% level: -2.8624
- 10% level: -2.5672
Test statistic (-14.29) is far below these critical values, which means we reject the null hypothesis that the series is non-stationary.

Conclusion:
The differenced time series is stationary and suitable for modeling.
