#### 1. Importing libraries and data

```python
# importing libraries and functions
import pandas as pd
import numpy as np
import os
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# assigning path for easier import of data
path = r'C:\Users\magia\NY CitiBike\2. Data\Prepared Data'
```

```python
# importing data
df = pd.read_csv(os.path.join(path, 'citibike_cleaned.csv'))
```

#### 2. Data Cleaning

```python
df.columns
```

```python
#dropping No column - not useful to analysis
df.drop(columns = {'bike_id', 'weekday', 'start_hour', 'start_time', 'start_station_id',
       'start_station_name', 'start_station_latitude',
       'start_station_longitude', 'end_time', 'end_station_id',
       'end_station_name', 'end_station_latitude', 'end_station_longitude', 'subscriber', 'gender',
       'birth_year_label', '88+_years_old'}, inplace = True)
```

```python
#renaming columns to more intuitive and concise names
df.rename(columns = {'trip_duration' : 'tdur', 
                     'birth_year' : 'byear'}, inplace = True)
```

```python
# checking NaNs in byear
df['byear'].isnull().sum()
```

```python
# dropping NaNs from birth_year for the purpose of our analysis

df = df.dropna(subset=['byear'])
```

```python
df = df[df['byear'].notnull()].copy()
```

```python
# changing birth year from floar to integers

df['birth_year'] = df['byear'].astype(int)
```

```python
df.dtypes
```

```python
df.drop(columns = {'birth_year'}, inplace = True)
```

```python
df.dtypes
```

```python
sns.histplot(df['tdur'], bins=15, kde = True) # Shows extreme values
```

#### 3. Data Prep for regression analysis

```python
# create scatterplot to see how the trip duration and birth year plot against each other
df.plot(x = 'byear', y='tdur', style ='o')
plt.ylabel('Trip duration')  
plt.show()
```

Age (here shown by birth year) does not seem to be a factor for people under 65 years old (the data is from 2013). We see the drip duration shorten for people born close to prior 1940 and sometime before 1950. 

```python
# Reshape the variables into NumPy arrays and put them into separate objects.

X = df['byear'].values.reshape(-1,1)
y = df['tdur'].values.reshape(-1,1)
```

```python
X
```

```python
y
```

```python
# Split data into a training set and a test set.

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

#### 4. Regression Analysis

```python
# Create a regression object.

regression = LinearRegression()
```

```python
# Fit the regression object onto the training set.

regression.fit(X_train, y_train)
```

```python
# Predict the values of y using X.

y_predicted = regression.predict(X_test)
```

```python
# Create a plot that shows the regression line from the model on the test set.

plot_test = plt
plot_test.scatter(X_test, y_test, color='gray', s = 15)
plot_test.plot(X_test, y_predicted, color='red', linewidth =3)
plot_test.title('CitiBike Trip duration by birth year (Test set)')
plot_test.xlabel('Birth year')
plot_test.ylabel('Trip duration')
plot_test.show()
```

```python
# Create objects that contain the model summary statistics.

rmse = mean_squared_error(y_test, y_predicted) # This is the mean squared error
r2 = r2_score(y_test, y_predicted) # This is the R2 score. 
```

```python
# Print the model summary statistics to  evaluate the performance of the model.

print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```

#### Slope: -1.97
For each additional year of age, trip duration decreases by ~2 seconds. The effect is very small, and possibly not practically meaningful.

#### R² score: 0.0017
This means the model explains only 0.17% of the variance in trip duration. In other words, age barely predicts trip duration — the relationship is extremely weak.

#### Mean Squared Error (MSE): ~275,000
As our trip duration is in seconds, this average squared error implies high variability in trip length.

```python
y_predicted
```

```python
# Create a dataframe comparing the actual and predicted values of y.

data = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_predicted.flatten()})
data.head(30)
```

#### We see that the predictions are often very off! The model might not be very useful.

#### Compare how the regression fits the training set

```python
# Predicting the X train

y_predicted_train = regression.predict(X_train)
```

```python
rmse = mean_squared_error(y_train, y_predicted_train)
r2 = r2_score(y_train, y_predicted_train)
```

```python
print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```

```python
# Visualizing the training set results.

plot_test = plt
plot_test.scatter(X_train, y_train, color='green', s = 15)
plot_test.plot(X_train, y_predicted_train, color='red', linewidth =3)
plot_test.title('CitiBike Trip Duration by Birth year(Train set)')
plot_test.xlabel('Birth year')
plot_test.ylabel('Trip duration')
plot_test.show()
```

#### Bonus: performance improvement after removing outliers

```python
# Clean the extreme values from the "birth year" variable observed during the consistency checks (in this case all birth years prior 1920, i.e. all people whose is higher than 93)

df = df[df['byear'] >=1920]
```

```python
# See how the scatterplot looks without outliers.

df.plot(x = 'byear', y='tdur', style='o')  
plt.title('Trip duration by birth year')  
plt.xlabel('birth year')  
plt.ylabel('trip duration')  
plt.show()
```

```python
# Reshape again.

X_2 = df['byear'].values.reshape(-1,1)
y_2 = df['tdur'].values.reshape(-1,1)
```

```python
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y_2, test_size=0.3, random_state=0)
```

```python
# Run and fit the regression.

regression = LinearRegression()  
regression.fit(X_train_2, y_train_2)
```

```python
# Predict.

y_predicted_2 = regression.predict(X_test_2)
```

```python
rmse = mean_squared_error(y_test_2, y_predicted_2)
r2 = r2_score(y_test_2, y_predicted_2)
```

```python
print('Slope:' ,regression.coef_)
print('Mean squared error: ', rmse)
print('R2 score: ', r2)
```

#### Slope: -2.249
For each additional year of age, trip duration decreases by about 2.25 seconds. This effect is quite small and likely not practically significant.

#### R² score: 0.0018
The model explains only 0.18% of the variance in trip duration, meaning age is a very weak predictor of trip duration. The low R² also indicates the model is not overfitting—it simply doesn't capture much of the underlying variation.

#### Mean Squared Error (MSE): ~272,000
Since trip duration is measured in seconds, this high MSE reflects a large spread in trip lengths, which is common in bike trip data

#### The small improvement after removing outliers shows some slight refinement but the core story stays the same: age alone isn’t a strong driver of trip duration in our bike trips data.

```python
# Visualizing the test set results.
plot_test = plt
plot_test.scatter(X_test_2, y_test_2, color='gray', s = 15)
plot_test.plot(X_test_2, y_predicted_2, color='red', linewidth =3)
plot_test.title('Trip duration by birth year (Test set)')
plot_test.xlabel('birth year')
plot_test.ylabel('trip duration')
plot_test.show()
```

```python
data = pd.DataFrame({'Actual': y_test_2.flatten(), 'Predicted': y_predicted_2.flatten()})
data.head(30)
```

#### Overall summary: Statistically, the model fits poorly, even when possible outlier birth data was removed. Age alone is a very weak predictor of trip duration — the data likely has many other stronger factors (e.g., trip type, weather, time of day, subscriber status). Not overfitted — just not useful on its own.
